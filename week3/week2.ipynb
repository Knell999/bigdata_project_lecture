```python
# 실행 시점 기준 1~100000 데이터를 수집
import requests
import json
import pandas as pd
API_KEY = '76744e525163726131303072746e6e6a'

def main() :
    data = pd.DataFrame()
    for j in range(1, 101):
        url = f'http://openapi.seoul.go.kr:8088/{API_KEY}/json/tbLnOpendataRtmsV/{(j - 1) * 1000 + 1}/{j*1000}'
        print(url)
        req = requests.get(url)
        content = req.json()
        con = content['tbLnOpendataRtmsV']['row']
        result = pd.DataFrame(con)
        data = pd.concat([data, result])
        
    data = data.reset_index(drop=True)
    data['DEAL_YMD'] = pd.to_datetime(data['DEAL_YMD'], format=("%Y%m%d"))
    data.to_csv('seoul_real_estate_million.csv', index=False, encoding='utf-8-sig')
    
    print(data.head())
    print(len(data), type(data))
if __name__ == "__main__":
    main()
```
